```python
import pandas as pd
import csv
print "Modules Printed"
```

```python
df_courses = pd.read_csv("ece_cs_courses.csv")
df_courses
```

```python
df_faculty = pd.read_csv("ece_faculty_netid_added.csv")
df_faculty
```

```python
count_courses = df_courses["Course"].count()
dict_courses = {}
for i in range(count_courses):
    course = df_courses["Course"][i]
    term = df_courses["Term"][i]
    netid = df_courses["NetId"][i]
    key = course+term
    dict_courses[key] = [course, term, netid]

# not needed
list_of_netid = []
for i in dict_courses.keys():
    netid = dict_courses[i][2]
    if netid not in list_of_netid:
        list_of_netid.append(netid)
dict_courses
```

```python
count_faculty = df_faculty["Netid"].count()
dict_faculty = {}
for i in range(count_faculty):
    netid = df_faculty["Netid"][i]
    firstname = df_faculty["FirstName"][i]
    lastname = df_faculty["LastName"][i]
    email = df_faculty["Email"][i]
    term = ""
    course_number = ""
    for j in dict_courses.keys():
        if dict_courses[j][2] == netid:
            term = dict_courses[j][1]
            course_number = dict_courses[j][0]
            key=netid+term+course_number
            dict_faculty[key] = [netid, firstname, lastname, email, term, course_number]
dict_faculty
```

```python
df3 = pd.DataFrame.from_dict(dict_faculty, orient='index', columns=['NetId', 'FirstName', 'LastName', 'Email', 'Term', 'Course Number'])
df3
```

```python
df3.to_csv('ece_cs_faculty_term_course_added.csv', index=False, encoding='utf-8')
```
