```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

print("Modules Loaded")
```

```python
webpage = requests.get('https://ece.illinois.edu/directory/grad-students.asp')
prefix = 'https://ece.illinois.edu/'
webpage_content = webpage.content
print(webpage_content)
```

```python
soup = BeautifulSoup(webpage_content, "html.parser")
grad_students_name_list = soup.find_all(attrs={'class':'name'})
# faculty = soup.find_all([attrs={'class':'name'}, "a"])
grad_students_netid_list = soup.find_all(attrs={'class':'email'})
print(grad_students_name)
```

```python
data = {}

for i in range(len(grad_students_name_list)):
    grad_students_name = grad_students_name_list[i]
    grad_students_netid = grad_students_netid_list[i]
    name = grad_students_name.a.string
    email = grad_students_netid.a.string
    data[name] = [name, email]
print(data)
```

```python
df = pd.DataFrame.from_dict(data, orient='index', columns=['Name', 'Email'])
df
```

```python
df.to_csv('ece_grad_students.csv', index=False, encoding='utf-8')
```
