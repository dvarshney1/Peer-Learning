```python
import pandas as pd
import csv
print ('Modules Loaded')
```

```python
df = pd.read_csv("ece_cs_courses.csv")
df
```

```python
count = df["Course"].count()
hash_map = {}

for i in range(count):
    value = df["Instructor"][i]
    key = df["Course"][i]
    term = df["Term"][i]
    key_ = key+term
    title = df["Title"][i]
    if key_ not in hash_map:
        hash_map[key_] = [key, value, term, title]
hash_map
```

```python
with open('ece_cs_courses.csv', 'rb') as f:
    file_ = csv.reader(f)
    list_of_name = map(tuple, file_)
list_of_name.pop(0)
print(list_of_name)
```

```python
map_ = {}
for line in list_of_name:
    name = line[1].split(',')
    lastname = name[0]
    first_and_middle = name[1].split(' ')
    firstname = first_and_middle[1]
#     print(firstname)
#     print(lastname)
    class_ = line[0]
    term = line[2]
    title = line[3]
    class_term = class_ + term
#     print(class_term)
    map_[class_term] = [class_, firstname, lastname, term, title]
print(map_)
```
