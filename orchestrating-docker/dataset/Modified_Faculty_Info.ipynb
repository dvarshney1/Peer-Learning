```python
import csv
import fileinput
import sys
import pandas as pd
```

```python
with open('ece_faculty.csv', 'rb') as f:
    file_ = csv.reader(f)
    list_of_name = map(tuple, file_)
list_of_name.pop(0)
print(list_of_name)
```

```python
map_ = {}
for line in list_of_name:
    netid_split = line[1].split('@')
    netid = netid_split[0]
    name_split = line[0].split(' ')
    length_ = len(name_split)
    firstname = name_split[0]
    lastname = name_split[length_ - 1]
    email = line[1]
    map_[netid] = [netid, firstname, lastname, email]
print(map_)
```

```python
df = pd.DataFrame.from_dict(map_, orient='index', columns=['Netid','FirstName','LastName','Email'])
df
```

```python
df.to_csv('ece_faculty_netid_added.csv', index=False, encoding='utf-8')
```
