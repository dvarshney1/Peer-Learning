```python
import pandas as pd
print ('Modules Loaded')
```

```python
df = pd.read_csv("gpa_dataset.csv")
df = df[df["Subject"] == "ECE"]
df = df[(df["Term"] == "Fall") | (df["Term"] == "Spring")]
df
```

```python
df = df.reset_index()
```

```python
df = df.drop(["index"], axis = 1)
```

```python
df = df.drop(["A+", "A", "A-", "B+", "B", "B-", "C+", "C-", "C", "D+", "D", "D-", "F"], axis=1)
df
```

```python
df2 = df.drop(["Year", "Term", "W"], axis=1)
df2
```

```python
df2['Mix'] = df2['Subject'] + "-" + df2['Number'].astype(str) 
df2
```

```python
count = df2["Number"].count()
hash_map = {}

for i in range(count):
    value = df2["Primary Instructor"][i]
    key = df2["Mix"][i]
    term = df2["YearTerm"][i]
    key_ = key+term
    title = df2["Course Title"][i]
    if key_ not in hash_map:
        hash_map[key_] = [key, value, term, title]
hash_map
```

```python
df3 = pd.DataFrame.from_dict(hash_map, orient='index', columns=['Course', 'Instructor', 'Term', 'Title'])
df3
```

```python
df3.to_csv('ece_cs_courses.csv', index=False, encoding='utf-8')
```
