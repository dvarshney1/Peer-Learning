Webscraping for ECE Fculty first


```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

print("Modules Printed")
```

```python
webpage = requests.get('https://ece.illinois.edu/directory/faculty.asp')
prefix = 'https://ece.illinois.edu/'
```

```python
webpage_content = webpage.content
print(webpage_content)
```

```python
soup = BeautifulSoup(webpage_content, "html.parser")
```

```python
faculty = soup.find_all(attrs={'class':'name'})
# faculty = soup.find_all([attrs={'class':'name'}, "a"])
faculty_netid = soup.find_all(attrs={'class':'email'})
print(faculty_netid)
```

```python
data = {}

for i in range(len(faculty)):
    each_faculty = faculty[i]
    each_faculty_netid = faculty_netid[i]
    name = each_faculty.a.string
    email = each_faculty_netid.a.string
    name.encode('utf-8')
    email.encode('utf-8')
    
#     print(email)
#     print(name)
    data[name] = [name, email]
```

```python
df = pd.DataFrame.from_dict(data, orient='index', columns=['Name', 'Email'])
print(df)
```

```python
df.to_csv('ece_faculty.csv', index=False, encoding='utf-8')
```
